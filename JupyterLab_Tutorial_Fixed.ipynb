# üìì Section 2: Create and Manage Files

## File Management Features

JupyterLab provides an enhanced file browser that allows you to:
- Create and organize files and folders
- Preview files without opening them
- Drag and drop files between folders
- Run scripts directly from the file browser

```python
import os
import datetime

# Create sample notebooks and files for demonstration
def create_sample_files():
    """Create sample files to demonstrate file management"""
    
    # Create a sample directory structure
    directories = [
        'sample_project',
        'sample_project/data',
        'sample_project/notebooks',
        'sample_project/scripts'
    ]
    
    for directory in directories:
        os.makedirs(directory, exist_ok=True)
        print(f"‚úÖ Created directory: {directory}")
    
    # Create sample files
    files = {
        'sample_project/README.md': '# Sample Project\n\nThis is a demonstration project.',
        'sample_project/requirements.txt': 'pandas\nnumpy\nmatplotlib\n',
        'sample_project/data/sample.csv': 'Name,Age,City\nJohn,25,NYC\nJane,30,LA\n',
        'sample_project/scripts/utils.py': 'def hello_world():\n    print("Hello from JupyterLab!")\n'
    }
    
    for filepath, content in files.items():
        with open(filepath, 'w') as f:
            f.write(content)
        print(f"‚úÖ Created file: {filepath}")

create_sample_files()
```

# üìä Section 3: Data Analysis Examples

## Sample Data Analysis Workflow

Let's demonstrate how JupyterLab excels at data analysis tasks:

```python
# Create a sample dataset for analysis
np.random.seed(42)

# Generate sample sales data
sales_data = {
    'Date': pd.date_range('2024-01-01', periods=100, freq='D'),
    'Product': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Headphones'], 100),
    'Sales_Amount': np.random.normal(1000, 300, 100),
    'Quantity': np.random.poisson(5, 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100)
}

df = pd.DataFrame(sales_data)
df['Sales_Amount'] = np.abs(df['Sales_Amount'])  # Ensure positive values

print("üìä Sample Sales Data Created:")
print(f"üìè Dataset shape: {df.shape}")
print("\nüîç First few rows:")
print(df.head())

print("\nüìà Basic Statistics:")
print(df.describe())
```

```python
# Create comprehensive visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('üìä Sales Data Analysis Dashboard', fontsize=16, y=1.02)

# Plot 1: Sales by Product
product_sales = df.groupby('Product')['Sales_Amount'].sum().sort_values(ascending=False)
axes[0, 0].bar(product_sales.index, product_sales.values, color='skyblue')
axes[0, 0].set_title('üí∞ Total Sales by Product')
axes[0, 0].set_ylabel('Sales Amount ($)')
axes[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Sales Trend Over Time
daily_sales = df.groupby('Date')['Sales_Amount'].sum()
axes[0, 1].plot(daily_sales.index, daily_sales.values, color='green', linewidth=2)
axes[0, 1].set_title('üìà Sales Trend Over Time')
axes[0, 1].set_ylabel('Daily Sales ($)')
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: Regional Sales Distribution
regional_sales = df.groupby('Region')['Sales_Amount'].sum()
colors = ['lightcoral', 'lightblue', 'lightgreen', 'gold']
axes[1, 0].pie(regional_sales.values, labels=regional_sales.index, autopct='%1.1f%%', 
               colors=colors, startangle=90)
axes[1, 0].set_title('üó∫Ô∏è Sales Distribution by Region')

# Plot 4: Quantity vs Sales Amount Scatter
scatter = axes[1, 1].scatter(df['Quantity'], df['Sales_Amount'], 
                           c=df['Product'].astype('category').cat.codes, 
                           alpha=0.6, s=50)
axes[1, 1].set_title('üîç Quantity vs Sales Amount')
axes[1, 1].set_xlabel('Quantity Sold')
axes[1, 1].set_ylabel('Sales Amount ($)')

plt.tight_layout()
plt.show()

print("\n‚ú® JupyterLab Visualization Benefits:")
print("   ‚Ä¢ Interactive plots with zoom and pan")
print("   ‚Ä¢ Multiple output formats (PNG, SVG, PDF)")
print("   ‚Ä¢ Side-by-side comparison capability")
print("   ‚Ä¢ Integrated with all major plotting libraries")
```

# ‚ö° Section 4: Performance Benchmarking

## JupyterLab Performance Analysis

Let's benchmark JupyterLab's performance with various data operations:

```python
# Performance benchmarking functions
def benchmark_data_processing():
    """Benchmark data processing operations"""
    print("üèÉ‚Äç‚ôÇÔ∏è Running Data Processing Benchmarks...")
    
    # Create large dataset
    size = 100000
    data = {
        'id': range(size),
        'value1': np.random.randn(size),
        'value2': np.random.randn(size),
        'category': np.random.choice(['A', 'B', 'C', 'D'], size)
    }
    
    # Benchmark DataFrame creation
    start_time = time.time()
    large_df = pd.DataFrame(data)
    creation_time = time.time() - start_time
    
    # Benchmark groupby operation
    start_time = time.time()
    grouped = large_df.groupby('category').agg({
        'value1': ['mean', 'std'],
        'value2': ['sum', 'max']
    })
    groupby_time = time.time() - start_time
    
    # Benchmark filtering
    start_time = time.time()
    filtered = large_df[large_df['value1'] > 0]
    filter_time = time.time() - start_time
    
    return {
        'dataset_size': size,
        'creation_time': creation_time,
        'groupby_time': groupby_time,
        'filter_time': filter_time
    }

# Run the benchmark
benchmark_results = benchmark_data_processing()

print(f"\nüìä Benchmark Results for {benchmark_results['dataset_size']:,} rows:")
print(f"   üìù DataFrame Creation: {benchmark_results['creation_time']:.3f} seconds")
print(f"   üë• GroupBy Operations: {benchmark_results['groupby_time']:.3f} seconds")
print(f"   üîç Data Filtering: {benchmark_results['filter_time']:.3f} seconds")
```

```python
# Memory usage monitoring
def get_memory_usage():
    """Get current memory usage statistics"""
    process = psutil.Process(os.getpid())
    memory_info = process.memory_info()
    
    return {
        'rss_mb': memory_info.rss / 1024 / 1024,  # Resident memory
        'vms_mb': memory_info.vms / 1024 / 1024   # Virtual memory
    }

memory_usage = get_memory_usage()

print("üß† Current Memory Usage:")
print(f"   Resident Memory: {memory_usage['rss_mb']:.1f} MB")
print(f"   Virtual Memory: {memory_usage['vms_mb']:.1f} MB")

# Performance comparison table
performance_comparison = {
    'Feature': ['Startup Time', 'Memory Usage', 'Large Dataset Handling', 
               'Multiple Files', 'Extension Loading', 'Rendering Speed'],
    'Classic Jupyter': ['Fast', 'Low', 'Good', 'Poor', 'Limited', 'Good'],
    'JupyterLab': ['Moderate', 'Higher', 'Excellent', 'Excellent', 'Rich', 'Excellent'],
    'Winner': ['Classic', 'Classic', 'JupyterLab', 'JupyterLab', 'JupyterLab', 'JupyterLab']
}

comparison_df = pd.DataFrame(performance_comparison)
print("\n‚öîÔ∏è Performance Comparison:")
print(comparison_df.to_string(index=False))
```

```python
# Visualization performance test
def benchmark_visualization():
    """Benchmark visualization rendering"""
    print("üìà Testing Visualization Performance...")
    
    # Create multiple plots
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    start_time = time.time()
    
    # Plot 1: Line plot
    x = np.linspace(0, 10, 1000)
    y = np.sin(x) * np.exp(-x/5)
    axes[0, 0].plot(x, y)
    axes[0, 0].set_title('Line Plot Performance')
    
    # Plot 2: Scatter plot
    x_scatter = np.random.randn(5000)
    y_scatter = np.random.randn(5000)
    axes[0, 1].scatter(x_scatter, y_scatter, alpha=0.5)
    axes[0, 1].set_title('Scatter Plot (5000 points)')
    
    # Plot 3: Histogram
    data_hist = np.random.normal(0, 1, 10000)
    axes[1, 0].hist(data_hist, bins=50, alpha=0.7)
    axes[1, 0].set_title('Histogram (10k samples)')
    
    # Plot 4: Heatmap
    heatmap_data = np.random.randn(50, 50)
    im = axes[1, 1].imshow(heatmap_data, cmap='viridis')
    axes[1, 1].set_title('Heatmap (50x50)')
    
    plt.tight_layout()
    
    render_time = time.time() - start_time
    
    plt.show()
    
    return render_time

viz_time = benchmark_visualization()
print(f"\n‚è±Ô∏è Visualization rendering time: {viz_time:.3f} seconds")
print("\nüí° JupyterLab Advantages for Visualizations:")
print("   ‚Ä¢ Interactive widgets work seamlessly")
print("   ‚Ä¢ Multiple output formats supported")
print("   ‚Ä¢ Side-by-side comparison of plots")
print("   ‚Ä¢ Integrated with Plotly, Bokeh, and other libraries")
```
